```r
data <- c(4, 3, 1, 7, 3, 2, 69)
summary(data)
```

```r
library(readxl)
library(ggplot2)
data <- read_excel("/content/Salary_Data.xlsx")
head(data)
```

```r
summary(data)
```

```r
par(pin = c(6, 3), mar = c(4, 0, 3, 0))  # Adjust margins: bottom, left, top, right
boxplot(data["Salary"],
        horizontal=TRUE,
        frame=FALSE,
        col='lightblue',
        main="Distribution of Salaries",
        xlab="Salaries")
```

```r
par(pin=c(6,4), mar = c(4,3,4,3))
hist(data[["Salary"]],
     main="Salary Distribution",
     xlab="Salary",
     col="coral")
```

```r
d <- density(data[["Salary"]])

par(pin=c(6,4), mar = c(4,3,4,3))
hist(data[["Salary"]],
     main="Salary Distribution",
     xlab="Salary",
     col="coral",
     prob=TRUE)
lines(d,col="blue", lwd=3)
```

```r
plot(d, col="blue", lwd=8)
polygon(d, col="lightblue", border="blue")
```
